```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create some data
t_values = np.linspace(0, 10, 11)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

for i, t in enumerate(t_values):
    Z = np.sin(np.sqrt(X**2 + Y**2) - t)
    # create an offset for each contour
    offset = i
    ax.contourf(X, Y, Z, zdir='y', offset=offset, alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.tight_layout()
plt.show()
```
