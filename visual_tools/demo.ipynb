```python
import numpy as np
import matplotlib.pyplot as plt

# some random signals 
def y(x):
    return 2 * np.exp(-3 * x**2) * np.sin(x) - np.cos(x) * np.sin(0.5 * x**2) + 2 * np.cos(x)**2 + 0.5

def y2(x):
    return 2 * np.exp(-1 * x**2) * np.cos(x) - np.sin(x) * np.cos(0.8 * x**2-x) + 2 * np.sin(x)**2 + 0.5
```

```python
t = np.linspace(0, 10, 6,endpoint=True)
t2 = np.linspace(0, 10, 21,endpoint=True)
t_ref = np.linspace(0, 10, 1001)

plt.figure(figsize=(10, 5))
plt.plot(t_ref, y(t_ref), 'k-')
plt.ylim(0, 4)
plt.xticks(np.arange(0, 10.5, 0.5))

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
# Hide y-axis
ax.get_yaxis().set_visible(False)
plt.savefig('demo_curve1.png', dpi=300)
```
